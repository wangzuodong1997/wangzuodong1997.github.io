#### Corrige

La suite de Fibonacci $F_n$ est définie par $F_0=0$, $F_1=1$ et $F_{n\geqslant 2} = F_{n-1} + F_{n-2}$. On montre $\underset{n \rightarrow \infty}{\mathrm{lim}} \left( \frac{F_{n+1}}{F_n}\right) = \varphi$ avec $\varphi =\frac{1+\sqrt{5}}{2}$ le nombre d'or.

````python
def next_fib_nb(X,Y):
    return X+Y

N=40
Fib_seq=np.zeros(N,dtype=int)
Fib_seq[1]=1
for i in range(2,N):
    Fib_seq[i]=next_fib_nb(Fib_seq[i-1],Fib_seq[i-2])

Ratio =Fib_seq[2:]/Fib_seq[1:-1]
#Fib_seq[2:] contient les termes de F_n de 2 à N (on exclut les deux premiers termes zéro et 1) : c'est F_(n+1)
#Fib_seq[1:-1] " de F_n de 1 à N-1 (on exclut le premier pour pouvoir diviser et le dernier) : c'est F_n
plt.plot(Ratio,'o',label='Numerical ratio')
plt.plot((1+np.sqrt(5))*.5*np.ones(N-2),'--',label='Golden ratio')
plt.xlabel("Integer n")
plt.ylabel("Ratio of consecutive Fibonacci numbers")
plt.title("Convergence toward the glolden ratio")
plt.legend()
plt.show()
print("On évalue le nombre d'or à ", Ratio[-1], " +/-",np.abs(Ratio[-1]-Ratio[-2]) )

plt.plot(np.abs(Ratio - (1+np.sqrt(5))*.5*np.ones(N-2)),'o',label='Numerical difference to the G. ratio')
plt.plot(np.exp(-1.0*np.arange(N-2)),'-',label='exp(-N)')
plt.xlabel("Integer n")
plt.ylabel("Difference to the golden ratio")
plt.title("Convergence toward the glolden ratio")
plt.yscale('log')
plt.legend()
plt.show()
````
